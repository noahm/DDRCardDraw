```python
import json
import re

data = []
diffLookup = {
    "N": "normal",
    "H": "hyper",
    "A": "another"
}
with open(r"C:\Users\telpi\Documents\GitHub\DDRCardDraw\iidx-motl5-base.txt", 'r', encoding='utf-8') as fp:
    for line in fp:
        tokens = line.split('\t')
        if len(tokens) <= 1:
            break

        if tokens[0] == '12H':
            lvl = 14
        elif tokens[0] == '12M':
            lvl = 13
        else:
            lvl = int(re.sub('\D', '', tokens[0]))
        name = tokens[1]
        diffClass = diffLookup[tokens[2]]
        artist = f"P1 {tokens[3]} | P2 {tokens[4]}"

        data.append({
            "name": name,
            "artist": artist,
            "bpm": "0",
            "jacket": "",
            "charts": [{
                "lvl": lvl,
                "style": "single",
                "diffClass": diffClass
            }]
        })

with open(r"C:\Users\telpi\Documents\GitHub\DDRCardDraw\iidx-motl5-base.json", 'w', encoding='utf-8') as fp:
    json.dump(data, fp)
```
